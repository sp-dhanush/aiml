## Write a Python program to analyze and visualize the data using NumPy and matplotlib Modules

```python
import pandas as pd # Weâ€™ll be using Pandas library to work with the dataset

from matplotlib import pyplot as plt #import pyplot from matplotlib as plt
```

```python
x= [1,2,3]
y=[1,4,9]
plt.plot(x,y)
plt.show()
```

```python
#add title
x= [1,2,3]
y=[1,4,9]
plt.plot(x,y)
plt.title('test plot')
plt.show()
```

```python
#add label
x= [1,2,3]
y=[1,4,9]
plt.plot(x,y)
plt.title('test plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#add third axis
x= [1,2,3]
y=[1,4,9]
z=[10,5,0]
plt.plot(x,y)
plt.plot(x,z)
plt.title('test plot')
plt.xlabel('x')
plt.ylabel('y and z')
plt.show()
```
